## 2. Timetables Data Exploration (Parsed from XML)

```python
import pandas as pd
df = pd.read_csv("../data/interim/timetables_parsed.csv")
print(df.shape)
print(df.isna().sum())
print(df['days_of_week'].value_counts().head(10))
df.head()
```

```python
wy_places = ['Leeds', 'Bradford', 'Wakefield', 'Huddersfield', 'Halifax', 
             'Dewsbury', 'Kirklees', 'Calderdale', 'Batley', 'Pudsey', 
             'Keighley', 'Brighouse', 'Castleford', 'Pontefract', 'Selby']

# quick check: how many rows have origin/destination matching known WY places
in_scope = df[df['origin'].str.contains('|'.join(wy_places), case=False, na=False) |
              df['destination'].str.contains('|'.join(wy_places), case=False, na=False)]

print(f"In-scope (West Yorkshire) rows: {in_scope.shape[0]} of {df.shape[0]}")
print(df['origin'].value_counts().head(20))
```

```python
# Known out-of-scope areas identified from the data (bordering counties, not West Yorkshire)
out_of_scope = ['Barnsley', 'Chesterfield', 'Rotherham', 'Harrogate', 'Blackburn', 
                 'Sheffield', 'Doncaster', 'Wombwell', 'Derby', 'Preston', 'Burnley']

# Known in-scope West Yorkshire boroughs/towns (Leeds, Bradford, Wakefield, Kirklees, Calderdale)
wy_places = ['Leeds', 'Bradford', 'Wakefield', 'Huddersfield', 'Halifax', 
             'Dewsbury', 'Batley', 'Pudsey', 'Keighley', 'Brighouse', 
             'Castleford', 'Pontefract', 'Bramley', 'Morley', 'Otley',
             'Ilkley', 'Shipley', 'Cleckheaton', 'Mirfield', 'Ossett']

# Exclude rows where EITHER origin or destination is clearly out of scope
mask_out = (df['origin'].str.contains('|'.join(out_of_scope), case=False, na=False) |
            df['destination'].str.contains('|'.join(out_of_scope), case=False, na=False))

# Keep rows where EITHER origin or destination matches a known WY place
mask_in = (df['origin'].str.contains('|'.join(wy_places), case=False, na=False) |
           df['destination'].str.contains('|'.join(wy_places), case=False, na=False))

df_wy = df[mask_in & ~mask_out].copy()

print(f"Filtered West Yorkshire rows: {df_wy.shape[0]} of {df.shape[0]}")
print(df_wy['origin'].value_counts().head(15))
print(df_wy['operator_name'].value_counts())
```

**Now: 3-month date expansion:**

```python
wyca_disruptions['Validity start'] = pd.to_datetime(wyca_disruptions['Validity start'], errors='coerce', utc=True)
wyca_disruptions['Validity end'] = pd.to_datetime(wyca_disruptions['Validity end'], errors='coerce', utc=True)

window_start = pd.Timestamp("2026-04-01", tz="UTC")
window_end = pd.Timestamp("2026-06-30", tz="UTC")

scoped_disruptions = wyca_disruptions[
    (wyca_disruptions['Validity start'] <= window_end) &
    ((wyca_disruptions['Validity end'].isna()) | (wyca_disruptions['Validity end'] >= window_start))
].copy()

print(scoped_disruptions.shape)
scoped_disruptions.head(10)
```

```python
print(wyca_disruptions['Validity start'].dtype)
print(wyca_disruptions['Validity start'].isna().sum())
print(wyca_disruptions['Validity start'].head(10))
```

```python
import re

# Reload untouched from the CSV filter (avoid double-parsing an already-corrupted column)
raw_disruptions = disruptions[disruptions['Organisation'].str.contains('WYCA|West Yorkshire', case=False, na=False)].copy()

def clean_and_parse_date(series):
    # Strip the trailing "(Coordinated Universal Time)" style text, keep the GMT offset
    cleaned = series.astype(str).str.replace(r'\s*\(.*\)$', '', regex=True)
    return pd.to_datetime(cleaned, errors='coerce', utc=True)

raw_disruptions['Validity start'] = clean_and_parse_date(raw_disruptions['Validity start'])
raw_disruptions['Validity end'] = clean_and_parse_date(raw_disruptions['Validity end'])

print("Parsed successfully:", raw_disruptions['Validity start'].notna().sum(), "of", len(raw_disruptions))
print(raw_disruptions['Validity start'].head(10))
```

```python
window_start = pd.Timestamp("2026-04-01", tz="UTC")
window_end = pd.Timestamp("2026-06-30", tz="UTC")

scoped_disruptions = raw_disruptions[
    (raw_disruptions['Validity start'] <= window_end) &
    ((raw_disruptions['Validity end'].isna()) | (raw_disruptions['Validity end'] >= window_start))
].copy()

print(scoped_disruptions.shape)
scoped_disruptions[['ID', 'Validity start', 'Validity end', 'Reason', 'Planned', 'Services affected']]
```
